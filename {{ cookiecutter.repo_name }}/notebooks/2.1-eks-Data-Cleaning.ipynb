```python
# Reads the csv file
import pandas as pd
df = pd.read_csv('lovedtracks.csv')
df.head(20)
```

```python
# Describes the raw dataset
df.describe()
```

```python
# Counts the number of missing values for each column
num_missing = df.isnull().sum()
print(num_missing)
```

```python
# Removes rows with missing values
new_df = df.dropna()
new_df.head(20)
```

```python
# Describes the modified dataset with missing values removed
new_df.describe()
```

```python
# Returns boolen values for duplicates
print(new_df.duplicated())
```

```python
# Removes duplicates (no changes in the dataframe)
df_copy = new_df.copy()
df_copy.drop_duplicates(inplace=True)
df_copy
```

```python
import numpy as np
from scipy.sparse import csr_matrix
from tqdm import tqdm

df_subset = df.head(2000)
df_reshaped = df_subset.drop('username', axis=1)
df_reshaped = df_reshaped.melt()

# 1 hot encoding of the values column to binary
df_encoded = pd.get_dummies(df_reshaped['value'])
df_encoded = df_encoded.astype(int)

# create a new column called id
df_encoded['id'] = np.tile(df_subset['username'].values, 50)

# group by id and sum the values
unique_ids = np.unique(df_encoded['id'])
num_ids = len(unique_ids)
num_columns = min(100000, df_encoded.shape[1])  # Specify the maximum number of columns

# Initialize the sparse matrix
sparse_data = []
rows = []
cols = []

# Iterate over unique ids with tqdm progress bar
for i in tqdm(range(num_ids), desc="Processing rows"):
    id_rows = df_encoded[df_encoded['id'] == unique_ids[i]].values[:, :-1]  # Exclude the last column
    id_sum = np.sum(id_rows, axis=0)
    non_zero_cols = np.nonzero(id_sum)[0]
    num_non_zero = len(non_zero_cols)
    
    # If the number of non-zero columns exceeds the desired maximum, randomly select the columns
    if num_non_zero > num_columns:
        non_zero_cols = np.random.choice(non_zero_cols, size=num_columns, replace=False)
    
    # Add the non-zero values and column indices to the sparse matrix
    sparse_data.extend(id_sum[non_zero_cols])
    rows.extend([i] * num_non_zero)
    cols.extend(non_zero_cols)

# Create the sparse matrix
sparse_matrix = csr_matrix((sparse_data, (rows, cols)), shape=(num_ids, num_columns))

# Convert the sparse matrix to a DataFrame
df_grouped = pd.DataFrame.sparse.from_spmatrix(sparse_matrix)
df_grouped['id'] = unique_ids

# Reset the column names
df_grouped.columns = ['lovedtrack_' + str(i) for i in range(num_columns)] + ['id']
```

```python
# Apply tqdm progress bar to the rows
tqdm.pandas(desc="Saving CSV")

# Create an empty DataFrame to store the concatenated rows
encoded_rows = pd.DataFrame()

# Iterate over the rows and concatenate them with progress bar
for _, row in tqdm(df_grouped.iterrows(), total=df_grouped.shape[0]):
    encoded_rows = pd.concat([encoded_rows, row.to_frame().T], ignore_index=True)

# Save the concatenated DataFrame to a CSV file
encoded_rows.to_csv('encoded.csv', index=False)
```
