```python
import pandas as pd

df = pd.read_csv('encoded.csv')
```

```python
# Sum up the values for each column (excluding the 'id' column)
column_sums = df.drop('id', axis=1).sum()

# Select columns with at least 10 ones
selected_columns = column_sums[column_sums >= 10].index.tolist()

# Include the 'id' column in the selected columns
selected_columns.append('id')

# Create a new DataFrame with the selected columns
new_df = df[selected_columns]

# Sort columns based on the number of ones in descending order (excluding the 'id' column)
sorted_columns = sorted(selected_columns[:-1], key=lambda col: new_df[col].sum(), reverse=True)

# Print the number of ones for each selected column in descending order
for column in sorted_columns:
    if column != 'id':
        num_ones = new_df[column].sum()
        print(f"Column '{column}' has {num_ones} ones.")
```
